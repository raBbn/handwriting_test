```
import numpy as np
import matplotlib.pyplot as plt
import pytesseract
try:
  from PIL import Image
except ImportError:
  import Image
```

```
filename = '1.PNG'
img = Image.open(filename)
custom_config = ' --oem 3 --psm 6'
extractedText = pytesseract.image_to_string(img, lang = 'nep', config = custom_config)
print(extractedText)
```

```
filename = 'download.png'
img = Image.open(filename)
custom_config = ' --oem 3 --psm 6'
extractedText = pytesseract.image_to_string(img, lang = 'eng', config = custom_config)
print(extractedText)
```

```
import cv2
img = cv2.imread('download.png')
plt.imshow(img)

```

```
# grayscale
def get_grayscale(image):
  return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```
img_gray = get_grayscale(img)
plt.imshow(img_gray)
```

```
#noise removal
def remove_noise(image):
  return cv2.medianBlur(image, 9)
```

```
img_noise = remove_noise(img)
plt.imshow(img_noise)
```

```
#dilation
def dilation(image):
  kernel = np.ones((5,5), np.uint8)
  return cv2.dilate(image, kernel, iterations = 1)
```

```
img_dilate = dilation(img_gray)
plt.imshow(img_noise)
```

```
#eroison
def erosion(image):
  kernel = np.ones((5,5), np.uint8)
  return cv2.erode(image, kernel, iterations = 2)
```

```
img_erode = erosion(img_gray)
plt.imshow(img_noise)
```
